```python
df2=df[['Customer ID', 'Product ID']]
```

```python
df2
```

```python
df3=df2.groupby('Customer ID').agg({'Product ID': pd.Series.tolist}).reset_index()
```

```python
df3
```

```python
key=[]
val=[]
jac_dict={}
for n in range(len(df3)):
    for i in range(len(df3)):
        
        if n<i:
            jac_sim=len(set(df3['Product ID'][n]).intersection(set(df3['Product ID'][i]))) / len(set(df3['Product ID'][n]).union(set(df3['Product ID'][i])))
            val.append(jac_sim)
            key.append([df3['Customer ID'][n],df3['Customer ID'][i] ])

for num in range (len(key)):
    jac_dict[tuple(key[num])]=val[num]
print(jac_dict)
```

```python
data=pd.DataFrame.from_dict(jac_dict, orient='index').reset_index()
```

```python
data
```

```python
new_col_list = ['Customer 1','Customer 2']
for n,col in enumerate(new_col_list):
    data[col] = data['index'].apply(lambda index: index[n])

data = data.drop('index',axis=1)
```
